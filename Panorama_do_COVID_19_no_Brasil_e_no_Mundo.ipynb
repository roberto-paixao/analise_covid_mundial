De acordo com o [Ministério da Saúde](https://coronavirus.saude.gov.br/sobre-a-doenca), os coronavírus são uma grande família de vírus comuns em muitas espécies diferentes de animais, incluindo camelos, gado, gatos e morcegos e raramente podem infectar humanos como foi o caso do MERS-CoV e SARS-CoV. A COVID-19 é uma doença infecciosa causada pelo recém-descoberto coronavírus SARS-CoV-2.

Transmitido principalmente por meio de gotículas provenientes de tosses ou espirros de pessoas infectadas, a gravidade dos sintomas varia muito de pessoa para pessoa.



Até o presente momento, observa-se que cerca de 80% dos casos confirmados são assintomáticos ou oligossintomáticos (poucos sintomas) e rápidos. A maioria das pessoas que se encaixam nesse grupo, se recuperam sem nenhuma sequela.

No entanto, cerca de 20% dos casos confirmados apresentam dificuldade respiratória e necessitam de assistência hospitalar dos quais 5% podem precisar de suporte ventilatório.

## Obtenção dos Dados

No ínicio da Pandemia, do final de 2019 para o primeiro semestre de 2020, pouco se sabia sobre a nova doença e o que se sabiam, muitas vezes eram informações conflitantes entre diversas fontes por se tratarem de pesquisas iniciais ou falta de organização dos dados, como por exemplo, a divergência entre os dados informados sobre a situação da Covid-19 pelos governos estaduais e federais e fontes não-oficiais o que faziam com que as análises não refletissem a realidade.

Agora, em Setembro de 2021, temos um conhecimento mais sólido sobre a nova doença e já temos vacinas desenvolvidas e sendo distribuídas ao redor do mundo.

Os dados mundiais foram obtidos através do portal [Our World in Data](https://ourworldindata.org/coronavirus), onde é atualizado diariamente com a data de novos casos confirmados, mortes, casos hospitalizados, testagem e vacinações e também outras variáveis de potencial interesse.

#### **Evolução dos casos mundiais**

<img src="https://image.freepik.com/vetores-gratis/mundo-com-19-covid-em-lupa_24877-62329.jpg" width=50%>



Vamos começar olhando os números de casos e óbitos decorrentes da Covid-19, mas usar os números absolutos de casos e óbitos não é uma boa maneira de analisar esses dados mas se usarmos a média móvel de 7 dias podemos ver as tendências de crescimento ou declínio de casos ou óbitos no decorrer do tempo.

A média móvel é calculada somando-se o número de casos/óbitos de cada um dos sete dias anteriores e dividindo esse resultado por 7. A taxa média de 7 dias suaviza as flutuações durante a semana e é um melhor indicador de tendência do que os números diários.

```
# Gráfico de méia móvel de novos casos

sns.set_style('ticks')
fig, ax = plt.subplots(figsize=(12,8))

novos_casos_suavizados = (df.loc[df.location == 'World', ['date', 'new_cases_smoothed']]
                          .reset_index(drop=True))

sns.lineplot(x='date', y='new_cases_smoothed',
             data=novos_casos_suavizados, ci=None, linewidth=3, color='DarkOrange')

ax.set_title('Média móvel (7 dias) de novos casos de COVID-19\n\n', color='Grey', size=20, loc='left')

ax.set_xlabel([], color='white')
ax.set_ylabel([], color='white')

ax.yaxis.set_ticklabels(['0', '0', '200K', '400K', '600K', '800K'])

sns.despine()
plt.tight_layout()
```

```
# Gráfico de méia móvel de novos óbitos

sns.set_style('ticks')
fig, ax = plt.subplots(figsize=(12,8))

novos_obitos_suavizados = (df.loc[df.location == 'World', ['date', 'new_deaths_smoothed']].reset_index(drop=True))

sns.lineplot(x='date', y='new_deaths_smoothed',
             data=novos_obitos_suavizados, ci=None, linewidth=3, color='DarkOrange')

ax.set_title('Média móvel (7 dias) de novos óbitos por COVID-19\n\n', color='Grey', size=20, loc='left')

ax.set_xlabel([], color='white')
ax.set_ylabel([], color='white')

ax.yaxis.set_ticklabels(['0', '0', '2K', '4K', '6K', '8K', '10K', '12K', '14K'])

sns.despine()
plt.tight_layout()
```

Olhando para a média móvel de novos casos e novos óbitos podemos ver as tendências durante o tempo e nos permite analisar melhor a situação da pandemia, é possível observar dois "picos" de novas infecções e um possível terceiro pico, da mesma forma acontece com os novos óbitos.

Porém, tentar gerar uma previsão de uma possível terceira onda sobre essas tendências pode não ser uma boa escolha visto que previsões tendem a mudar muito por conta do comportamento das pessoas, por exemplo:

* **Se os números de casos novos voltarem a aumentar, as pessoas vão começar a usar as máscaras e aplicar o distanciamento social novamente até que elas percebam que os casos estão diminuindo para então começar a relaxar nas medidas preventivas novamente.**

#### **Taxa de Reprodução** (R0)

A **taxa de reprodução** (R0) do vírus nos informa a velocidade em que o vírus se propaga pela população. O **R0** é a média de pessoas que pegarão a doença de uma **única** pessoa infectada em uma população que nunca viu a doença antes. Isso quer dizer que, segundo a [OxfordBrasil](https://oxfordbrazilebm.com/index.php/quando-isso-acabara-uma-introducao-aos-numeros-de-reproducao-viral-r0-e-re/):

> _Se a média do R0 na população for maior que 1, a infecção se espalhará exponencialmente. Se o R0 for menor que 1, a infecção se espalhará apenas lentamente, e eventualmente desaparecerá. Quanto maior o valor de R0, mais rapidamente uma epidemia progredirá._

Vamos agora analisar a taxa de reprodução do vírus nos países.

#### **Número de Testes vs. Taxa de Positividade**



Os números de casos são contabilizados quando o resultado de um teste é **positivo**.

Os números de testes flutuam bastante devido ao local onde os testes são feitos e a quando os laboratórios reportam o resultado dos testes pois um surto repentino de testes pode refletir em um número de testes feitos em um único dia em grupos específicos como casas de repouso ou prisões e entre outros durante um surto, o fato de que alguns laboratórios e hospitais reportam os resultados de testes apenas em dias da semana então é comum que os números de caiam durante os fins de semanas. 

A taxa de positividade indica o quão difícil ou fácil será encontrar um caso, no qual reflete tanto a disseminação e o quão generalizado os testes são.

* **Exemplo**: _Se uma taxa de positividade for 20%, significa que não precisaremos procurar muito para encontrar um caso, quando a taxa for 1% significa que deverá ser feito muitos testes para encontrar um caso positivo._


> **Quanto mais a COVID-19 espalhar, maior será a taxa de positividade**

Entretanto, deve-se ater ao contexto, pois se uma taxa de positividade for 60%, pode nos indicar que os testes estejam sendo feitos apenas em hospitais devido a falta de testes e é onde os casos serão mais óbvios e não estão sendo feitos na população em geral onde os casos sejam mais leves.

Após analisar os dados obtidos sobre a situação mundial e brasileira, podemos concluir que:

* A alta taxa de contágio em alguns países podem ser devido ao surgimento de novas variantes do coronovíruas ou por medidas de flexibilização mais brandas.
